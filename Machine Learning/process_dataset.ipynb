```python
import json
import pandas as pd
```

```python
df = pd.read_json('News_Category_Dataset_v3.json', lines=True)
df.head()
```

```python
df.category.unique()
```

```python
categories = {
    'TECH': 'Technology',
    'ENVIRONMENT': 'Environment',
    'ENTERTAINMENT': 'Entertainment',
    'POLITICS': 'Politics',
    'EDUCATION': 'Education',
    'CRIME': 'Crime',
    'SPORTS': 'Sports',
    'BUSINESS': 'Business',
    'TRAVEL': 'Travel',
    'MONEY': 'Money'
}
df = df.loc[df['category'].isin(categories)]
df
```

```python
for category in categories:
    df.loc[df['category']==category, 'category'] = categories[category]
df.loc[df['category']=='Entertainment']
```

```python
df = df[['headline', 'short_description', 'category']]
```

```python
df.count()
```

```python
df.isna().any()
```

```python
df= df.rename(columns={'headline':'title', 'short_description':'description'})
```

```python
df.to_csv('news_category_dataset.csv', index=False)
```
